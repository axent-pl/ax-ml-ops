```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_train, df_test = load_and_transform_data()
X_cols, y_col = get_xy_cols(df_train)
```

## Plot distributions

```python
cols = sorted(X_cols)
fig, axes = plt.subplots(len(cols), 3, figsize=(16,4*len(cols)), constrained_layout=True)
for idx, col in enumerate(cols):
    sns.kdeplot(data=df_train, ax=axes[idx][0], x=col, fill=True).set_title(f'Distribution of "{col}"')
    sns.kdeplot(data=df_train, ax=axes[idx][1], x=col, fill=True, palette='RdYlGn' , hue=y_col).set_title(f'Distribution of "{col}" by {y_col}')
    sns.boxplot(data=df_train, ax=axes[idx][2], x=col, palette='RdYlGn', y=y_col, orient='h').set_title(f'BoxPlot of "{col}" by {y_col}')
plt.show()
```
